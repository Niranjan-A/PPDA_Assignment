## **4. Combining & Merging Datasets**

#### *4.1. Using concat()*


```python
data_1 = {"Member Id":[401,402,403,404,405],
        "Member Name":["Yami", "Asta", "Noelle Silva", "Luck Voltia", "Vanessa Enoteca"],
        "Magic Type":["Dark Magic", "Demon Magic", "Water Magic", "Lightning Magic","Witch Spells"]}
  

data_2 = {"Member Id":[406,407,408,409,410,411],
         "Member Name":["Henry Legolant", "Magna Swing", "Grey", "Finral Roulacase", "Gordon Agrippa", "Gauche Adlai"],
          "Magic Type" :["Demon Magic", "Fire Magice", "Transformation Magic", "Spatial Magic", "Curse Magic", "Mirror Magic"]}
          

df1= pd.DataFrame(data_1, index=[0, 1, 2, 3, 4])
df1
```

```python
df2 = pd.DataFrame(data_2, index=[5, 6, 7, 8, 9, 10])
df2
```

```python
frames = [df1, df2]
final_df = pd.concat(frames)

final_df
```

#### *4.2. Using merge()*

```python
data_a = {"Member Id":[401,402,403,404,405,406,407,408,409,410,411],
        "Member Name":["Yami", "Asta", "Noelle Silva", "Luck Voltia", "Vanessa Enoteca", "Henry Legolant", "Magna Swing", "Grey", "Finral Roulacase", "Gordon Agrippa", "Gauche Adlai"]}

data_b = {"Member Id":[401,402,403,404,405,406,407,408,409,410,411],
          "Magic Type" :["Dark Magic", "Demon Magic", "Water Magic", "Lightning Magic","Witch Spells", "Demon Magic", "Fire Magice", "Transformation Magic", "Spatial Magic", "Curse Magic", "Mirror Magic"]}

df_a = pd.DataFrame(data_a)
df_a
```

```python
df_b = pd.DataFrame(data_b)
df_b
```

```python
df_a.merge(df_b, left_on='Member Id', right_on='Member Id')
```
